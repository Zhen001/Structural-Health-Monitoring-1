# 下载csv数据

```python
import requests
import time
for year in range(2001,2020):
    url = 'http://www.data.jma.go.jp/fcd/yoho/typhoon/position_table/table%s.csv'%year
    download_path = r'E:\【论文】\【大论文】\气象资料收集\数据\日本气象厅台风数据\台风数据 %s.csv'%year
    req = requests.get(url)
    open(download_path, 'wb').write(req.content)
    time.sleep(1)
```

# 整理并保存成pkl

```python
import os
import pickle
import numpy as np
import pandas as pd

main_path = r'E:\【论文】\【大论文】\气象资料收集\数据\日本气象厅台风数据'
file_list = ['%s\\%s'%(main_path,x) for x in os.listdir(main_path) if x.split('.')[-1]=='csv']
direction_dict = {1：'东北',2：'东',3：'东南',4：'南',5：'西南',6：'西',7：'西北',8：'北'}

for file in file_list:
    with open(file,'r', encoding='shift-jis') as f:
        data = pd.read_csv(f, engine='python')
    data.index = data.apply(lambda x: pd.Timestamp(x['年'],x['月'],x['日'],x['時（UTC）']), axis=1) + pd.Timedelta('8H')
    data.drop(columns=['年', '月', '日', '時（UTC）', '上陸'], inplace=True)
    data.columns = ['台风代号', '台风名', '等级', '纬度', '经度', '中心气压(百帕)', '最大风速(m/s)', '50KT长径方向', '50KT长径(km)', '50KT短径(km)', '30KT长径方向', '30KT长径(km)', '30KT短径(km)']
    for 台风代号 in {x for x in data['台风代号'].values}:
        typhoon_tracks = data[data['台风代号']==台风代号]
        台风名 = typhoon_tracks.iloc[0,1].capitalize()
        typhoon_tracks.loc[:,'台风名'] = 台风名
        typhoon_tracks.loc[:,'最大风速(m/s)'] *= 1852/3600 # 节/(m/s)转换
        typhoon_tracks.loc[:,'50KT长径(km)'] *= 1.852 # 海里/km转换
        typhoon_tracks.loc[:,'50KT短径(km)'] *= 1.852 # 海里/km转换
        typhoon_tracks.loc[:,'30KT长径(km)'] *= 1.852 # 海里/km转换
        typhoon_tracks.loc[:,'30KT短径(km)'] *= 1.852 # 海里/km转换
        typhoon_tracks.loc[:,'50KT长径方向'] = [direction_dict.get(x) for x in typhoon_tracks.loc[:,'50KT长径方向']]
        typhoon_tracks.loc[:,'30KT长径方向'] = [direction_dict.get(x) for x in typhoon_tracks.loc[:,'30KT长径方向']]
        
        file_path = r'E:\【论文】\【大论文】\气象资料收集\数据\日本气象厅台风数据\%s %s(日本气象厅).pkl'%(台风代号,台风名)
        with open(file_path, 'wb') as f:
            pickle.dump(typhoon_tracks, f)
```
